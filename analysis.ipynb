```python
import os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import json
```

```python
path = "./results/cifar10_brightness_cnt=5/"

ans = []
for file_path in os.listdir(path):
    if (file_path.endswith(",results.json")):
        
        temp = {}
        for item in file_path.split(","):
            if (item.find("=") != -1):
                key, value = item.split("=")
                temp[key] = value

        with open(os.path.join(path, file_path), "r") as f:
            data = json.loads(f.read())

            temp["test_auc"] = data["test_auc"]
            if "test_auprc" in data:
                temp["test_auprc"] = data["test_auprc"]

        ans.append(temp)

ans = pd.DataFrame(ans,dtype=float)
```

```python
ans = ans.dropna().reset_index(drop=True)
ans = ans.sort_values(by="test_auprc",ascending=False).reset_index(drop=True)
```
